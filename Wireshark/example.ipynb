```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('example.csv')
data.drop('Info', inplace=True, axis=1)

plt.figure(figsize=(10,6))
x=data['Protocol'].value_counts()[0:8] 
other_sum=sum(data['Protocol'].value_counts()[9:])
x.at['Other']=other_sum 
labels=x.index
explode = np.append([0.05],np.zeros(len(x)-1)) 
plt.pie(x,labels=labels,explode=explode,autopct='%.02f%%')
plt.title('Detail Of Protocol Info', fontsize=20)
```

```python
plt.figure()
counts, _, patches=plt.hist(data['Protocol'],
        bins = np.arange(len(data['Protocol'].value_counts())) - 0.5,
        rwidth=0.5)

for count, patch in zip(counts,patches):
    plt.text(x=patch.get_x()+0.05, y=patch.get_height()+50,s=str(int(count)))

plt.xticks(rotation=40)
plt.ylabel('Packet Numbers',fontsize=20) 
plt.xlabel('Protocol Names', fontsize=20)
plt.show()
```

```python
plt.figure()
plt.plot(data.groupby(by='Time')['Length'].sum().index,data.groupby(by='Time')['Length'].sum())

plt.xticks(rotation=40)
plt.ylabel('Time(seconds)', fontsize=20)
plt.xlabel('Packets Length' ,fontsize=20)
plt.show()
```
